```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_z_slice(file_path, slice_idx):
    # Load the 3D realization
    data = np.load(file_path)
    
    # Basic validation of dimensions
    if data.ndim != 3:
        raise ValueError(f"Expected 3D data, but got {data.ndim}D.")

    # Select the Z-slice
    # Assuming the shape is (Z, Y, X)
    z_slice = data[slice_idx, :, :]

    # Plotting
    plt.figure(figsize=(8, 6))
    plt.imshow(z_slice, cmap='viridis', origin='lower')
    plt.colorbar(label='Intensity/Value')
    plt.title(f"Z-Slice Visualization (Index: {slice_idx})")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()

# Usage
visualize_z_slice('realization_03.npy', slice_idx=10)
```
