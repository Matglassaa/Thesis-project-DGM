```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_z_slice(file_path, slice_idx):
    # Load the 3D realization
    data = np.load(file_path)
    
    # Basic validation of dimensions
    if data.ndim != 3:
        raise ValueError(f"Expected 3D data, but got {data.ndim}D.")

    # Select the Z-slice
    # Assuming the shape is (Z, Y, X)
    z_slice = data[slice_idx, :, :]

    # Create a figure with 1 row and 2 columns
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    # --- Plot 1: Z-Slice Visualization ---
    im = axes[0].imshow(z_slice, cmap='viridis', origin='lower')
    fig.colorbar(im, ax=axes[0], label='Intensity/Value')
    axes[0].set_title(f"Z-Slice Visualization (Index: {slice_idx})")
    axes[0].set_xlabel("X-axis")
    axes[0].set_ylabel("Y-axis")
    
    # --- Plot 2: Histogram of Cell Values ---
    # .ravel() flattens the 2D array into 1D for the histogram
    axes[1].hist(z_slice.ravel(), bins=50, color='skyblue', edgecolor='black')
    axes[1].set_title(f"Distribution of Values (Slice {slice_idx})")
    axes[1].set_xlabel("Intensity/Value")
    axes[1].set_ylabel("Frequency (Number of Cells)")
    axes[1].grid(axis='y', alpha=0.75)

    # Adjust layout so plots don't overlap
    plt.tight_layout()
    plt.show()

# Usage
visualize_z_slice('realization_01.npy', slice_idx=30)
```
