```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_slice(file_path, slice_idx):
    # Load the 3D realization
    data = np.load(file_path)
    
    # Basic validation of dimensions
    if data.ndim != 3:
        raise ValueError(f"Expected 3D data, but got {data.ndim}D.")

    # Select the Z-slice
    # Assuming the shape is (Z, Y, X)
    x_slice = data[:, slice_idx, :]
    y_slice = data[:, :, slice_idx]
    z_slice = data[slice_idx, :, :]

    # Plotting
    fight, axs = plt.subplots(1, 3, figsize=(18, 6))
    axs[0].imshow(x_slice, cmap='viridis', origin='lower')
    axs[0].set_title(f"X-Slice (Index: {slice_idx})")
    axs[0].set_xlabel("X-axis")
    axs[0].set_ylabel("Z-axis")

    axs[1].imshow(y_slice, cmap='viridis', origin='lower')
    axs[1].set_title(f"Y-Slice (Index: {slice_idx})")
    axs[1].set_xlabel("Y-axis")
    axs[1].set_ylabel("Z-axis")

    axs[2].imshow(z_slice, cmap='viridis', origin='lower')
    axs[2].set_title(f"Z-Slice (Index: {slice_idx})")
    axs[2].set_xlabel("X-axis")
    axs[2].set_ylabel("Y-axis")
    
    plt.suptitle("Visualization of 3D Realization Slices", fontsize=16)
    plt.tight_layout()
    plt.show()

    plt.hist(data.flatten(), bins=50, color='blue', alpha=0.7)
    plt.title("Histogram of Realization Values")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

# Usage
visualize_slice('realization_05.npy', slice_idx=10)
```
